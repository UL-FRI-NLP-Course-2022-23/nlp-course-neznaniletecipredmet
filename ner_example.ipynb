```python
people = []
candidates = []
for s in doc.iter_tokens():
    if(s.words[0].upos  == "NOUN"):
        candidates.append(s)
    elif(s.words[0].upos == "PROPN"):
        people.append(s)
```

```python
candidates
```

```python
people_names = [p.lemma for p in people]
unique = []
for name in people_names:
    already = False
    for unique_name in unique:
        if(ls.ratio(name, unique_name) > 0.9):
            already = True
            break
    if(not already):
        unique.append(name)
```
